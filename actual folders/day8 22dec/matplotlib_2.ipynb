```python
lista = [1, 2, 3, 4]
listb = [1, 3, 7, 13]
listc = [16, 13, 9, 5]
```

```python
fig, ax = plt.subplots()

bx = ax.twinx() #mirror 

ax.plot(lista, listb, 'g-')
bx.plot(lista, listc, 'b-')

ax.set_xlabel('days')
ax.set_ylabel('temperature')
bx.set_ylabel('winds')

plt.show()
```

```python
fig, ax = plt.subplots()

bx = ax.twiny() # mirror 

ax.plot(lista, listb, 'g-')
bx.plot(lista, listc, 'b-')

ax.set_xlabel('days')
ax.set_ylabel('temperature')
bx.set_ylabel('winds')

plt.show()
```
