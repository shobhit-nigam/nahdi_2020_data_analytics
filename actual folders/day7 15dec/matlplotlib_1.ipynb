```python
lista = [1, 2, 3, 4]
listb = [1, 3, 7, 13]
listc = [1, 5, 9, 16]
```

```python
plt.plot(lista, listb, "go")   
plt.plot(lista, listc, "rd")  

plt.xlabel("hours")
plt.ylabel("productivity")
plt.title("tuesday plot")

plt.show()
```

```python
# lista vs listb
plt.plot(lista, listb, "go") 
plt.plot(lista, listb, "y:")

# lista vs listc
plt.plot(lista, listc, "rd")
plt.plot(lista, listc, "b--")

plt.xlabel("hours")
plt.ylabel("productivity")
plt.title("tuesday plot")

plt.show()
```

```python
listd = [1, 4, 9, 16]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

ax[0, 0].plot(lista, listb, 'g-')
ax[0, 1].plot(lista, listc, 'r--')
ax[1, 0].plot()
ax[1, 1].plot(lista, listd, 'c:')

#optional 
ax[0, 0].set_title("monday")
ax[0, 1].set_title("tuesday")
ax[1, 0].set_title("wednesday")
ax[1, 1].set_title("thursday")


plt.show()
```
